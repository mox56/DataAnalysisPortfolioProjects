```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plot we will create

#Read in the data
df= pd.read_csv(r'C:\Users\lenovo\data analysis\Movie\movies.csv')
```

```python
df.head()
```

```python
#Check if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df =df.dropna()
```

```python
#Data typr for our columns
df.dtypes
```

```python
#Change data types of columns for better viewing


df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#Create correct year column

df['YearCorrect'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head(10)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates

df.drop_duplicates()
```

```python
#Scatter plot with budget vs gross

plt.scatter(x= df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for file')

plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Looking at correlation
df.corr(numeric_only=True, method='pearson') #Pearson , Kendall, Spearman
```

```python
#High correlation between budget and gross
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric  features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')
plt.show
```

```python
# Look at company correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
#Checking if its numrized correctly
df.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric  features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')
plt.show
```
